```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.io as pio
```

```python
df=pd.read_csv("./dataset/car_prices.csv")
```

```python

df.head()
```

```python
df['saledate'] = pd.to_datetime(df['saledate'], errors='coerce', utc=True)
```

```python
print(df[['saledate']].head())
```

```python
df = df.dropna(subset=['saledate'])
```

```python
df['year'] = df['saledate'].dt.year
df['month'] = df['saledate'].dt.month
```

```python
total_sales = df['sellingprice'].sum()
number_of_orders = len(df)
```

```python
print(f"Total Sales: ${total_sales:.2f}")
print(f"Number of Orders: {number_of_orders}")

# Plot total sales over years
sales_by_year = df.groupby('year')['sellingprice'].sum()
sales_by_year.plot(kind='bar', color='skyblue', figsize=(10,5))
plt.title('Total Sales by Year')
plt.xlabel('Year')
plt.ylabel('Sales ($)')
plt.show()
```

```python
avg_revenue_per_order = df['sellingprice'].mean()
print(f"Average Revenue per Order: ${avg_revenue_per_order:.2f}")

# Optional histogram
plt.figure(figsize=(6,4))
sns.histplot(df['sellingprice'], bins=30, kde=True, color='orange')
plt.title('Distribution of Revenue per Order')
plt.xlabel('Revenue ($)')
plt.show()
```

```python
top5_qty = df['model'].value_counts().head(5)
print("Top 5 Products by Quantity Sold:\n", top5_qty)

# By revenue
top5_revenue = df.groupby('model')['sellingprice'].sum().sort_values(ascending=False).head(5)
print("Top 5 Products by Revenue:\n", top5_revenue)
```

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
top5_qty.plot(kind='bar', color='green')
plt.title('Top 5 Products by Quantity Sold')
plt.ylabel('Number of Orders')

plt.subplot(1,2,2)
top5_revenue.plot(kind='bar', color='purple')
plt.title('Top 5 Products by Revenue')
plt.ylabel('Revenue ($)')

plt.tight_layout()
plt.show()
```

```python
monthly_sales = df.resample('M', on='saledate')['sellingprice'].sum()

plt.figure(figsize=(12,5))
monthly_sales.plot(color='red')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales ($)')
plt.show()
```

```python
time_corr = df.groupby(['year','month'])['sellingprice'].sum().reset_index().corr()
print("Correlation matrix (year, month vs sales):\n", time_corr)
```

```python
peak_months = df.groupby(df['saledate'].dt.month)['sellingprice'].sum().sort_values(ascending=False)
print("Peak Sales Months:\n", peak_months)
```

```python
peak_months.plot(kind='bar', color='teal', figsize=(10,5))
plt.title('Sales by Month')
plt.xlabel('Month')
plt.ylabel('Revenue ($)')
plt.show()
```
